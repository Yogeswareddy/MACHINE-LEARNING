```python
from imblearn.over_sampling import RandomOverSampler
```

```python
oversample = RandomOverSampler(sampling_strategy='minority')
```

```python
X_over, y_over = oversample.fit_resample(X, y)
```

```python
new_df=pd.concat((X_over,y_over),axis=1)
new_df
```
